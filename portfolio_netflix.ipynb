```python
def genre(LIST):
    category = []
    list = []

    len_listed = len(LIST)

    for i in range(len_listed):
        list = LIST[i].split(",")
        len_list = len(list)
        for j in range(len_list):
            value = list[j].strip()
            if value not in category:
                category.append(value)

    df_c = pd.DataFrame(category)
    df_c.insert(1,'count', 0 * len(df_c[0]))
    df_c.rename(columns={0:'content'}, inplace=True)

    len_df_c = len(df_c['content'])      

    for i in range(len_listed):
        list = LIST[i].split(",")
        len_list = len(list)
        for j in range(len_list):
            value = list[j].strip()
            for k in range(len_df_c):
                if value == df_c['content'][k]:
                    df_c['count'][k]+=1
    
    return df_c
```

```python
df_c = genre(netflix['listed_in'])
```

```python
print(df_c[df_c['content']=='International Movies']['count']/sum(df_c['count']))
print(df_c[df_c['content']=='Dramas']['count']/sum(df_c['count']))
print(df_c[df_c['content']=='Comedies']['count']/sum(df_c['count']))
```

```python
len_dict = len(df_c['content'])
content_dict = {}

for i in range(len_dict):
    a = df_c['content'][i]
    b = df_c['count'][i]
    content_dict[a] = b
```

```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt
import numpy as np
from PIL import *

cand_mask=np.array(Image.open('D:/circle.png'))

words = content_dict

wordcloud = WordCloud(
    font_path = 'malgun.ttf',
    background_color='black',
    colormap='Greens',
    mask=cand_mask,
).generate_from_frequencies(words)

plt.figure(figsize=(5,5))
plt.imshow(wordcloud)
plt.axis('off')
```
